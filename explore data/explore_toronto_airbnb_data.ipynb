```python
import pandas as pd
```

### Large Listings Dataset

```python
# Large Airbnb listings dataset url
listings_url = "https://data.insideairbnb.com/canada/on/toronto/2025-11-11/data/listings.csv.gz"

# 1. Read just the first 5 rows to 'peek'
# Pandas handles the .gz compression automatically!
listings_df = pd.read_csv(listings_url, nrows=5)

# 2. Look at the column names (there are over 70!)
print(f"Total Columns: {len(listings_df.columns)}")
print(listings_df.columns)

# 3. View the data
listings_df.head()
```

```python
listings_df.info()
```
